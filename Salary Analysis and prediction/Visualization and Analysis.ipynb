```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cleaned_data.csv')
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='blue')

# Adding titles and labels
plt.title('Age Distribution of Individuals', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.show()
```

- Majority of employees are 25-35. Means the employees are energetic and can be suitable for long run

```python
# Counting the occurrences of each education level
education_counts = df['Education Level'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=education_counts.index, y=education_counts.values, palette='viridis')

plt.title('Education Level Distribution', fontsize=16)
plt.xlabel('Education Level', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.show()
```

- Most of the employees have bachelors degree in the company. High school is the lowest

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])

plt.title('Gender Distribution', fontsize=14)

plt.show()
```

- Majority of the employees are males with 54 percent.

```python
job_title_counts = df['Job Title'].value_counts()

plt.figure(figsize=(12, 8))
sns.barplot(x=job_title_counts.values, y=job_title_counts.index, palette='crest')

plt.title('Job Title Distribution', fontsize=16)
plt.xlabel('Frequency', fontsize=14)
plt.ylabel('Job Title', fontsize=14)

plt.show()
```

- Based on this observation, one could hypothesize that roles such as Software Developer, Data Analyst/Scientist, and Manager/Director are in higher demand compared to other job titles. This also suggests that positions in Finance/Accounting, Operations/Supply Chain Management, and Customer Service are less in demand and may offer lower compensation.

```python
job_title_counts = df['Race'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=job_title_counts.values, y=job_title_counts.index, palette='viridis')

plt.title('Race Distribution', fontsize=14)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Race', fontsize=12)

plt.show()
```

- This graph help us to know about the racial distribution in the dataset. From the graph, it is clear that most of the employees are either White or Asian, followed by Korean, Chinese, Australian and Black.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Salary', data=df, palette='Set2')
plt.title('Salary by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Salary', fontsize=14)
plt.show()
```

-  In the boxplot the employees from Other gender has quite high salary as compared to Makes and Females.
-  median salary of others is quite high maybe because of the less data availble for that section

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education Level', y='Salary', data=df, palette='Set3')
plt.title('Salary by Education Level', fontsize=16)
plt.xlabel('Education Level', fontsize=14)
plt.ylabel('Salary', fontsize=14)

plt.show()
```

- PhD holders have the highest median salary, followed by those with Master's and Bachelor's degrees, while employees without a degree have the lowest median salary. Bachelor's degree holders have salaries centered around 50,000 $$, while employees without a degree have salaries mainly between $40,000$ and $45,000.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Years of Experience', y='Salary', data=df, palette='Set2')
plt.title('Salary by Years of Experience', fontsize=16)
plt.xlabel('Years of Experience', fontsize=14)
plt.ylabel('Salary', fontsize=14)
plt.show()
```

- From this scaaterplot, it is clear that on the whole, the salary of the employees is increasing with the years of experience. However, on closer look we can see that similar experience have different salaries. This is because the salary is also dependent on other factors like job title, age, gender education level as discussed earlier.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Education Level', y='Salary', data=df, estimator=lambda x: sum(x)/len(x), palette='Set3')
plt.title('Average Salary by Education Level', fontsize=16)
plt.xlabel('Education Level', fontsize=14)
plt.ylabel('Average Salary', fontsize=14)

plt.show()
```

- It is not a surprise that the phd level of education boost the salary level of employees. Higher the education, Higher the salary

```python
plt.figure(figsize=(10, 7))
sns.barplot(x='Salary', y='Country', data=df, estimator=lambda x: sum(x)/len(x), palette='Set3')
plt.title('Average Salary by Country', fontsize=16)
plt.xlabel('Average Salary', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.show()
```

- The distribution of salary is almost same for all of the employees from different country
